#### Concatenate

```python
india_weather=pd.DataFrame({
    'city':['mumbai','bhubaneswar','bangalore'],
    'temperature':[30,32,25],
    'humidity':[55,65,70]})
india_weather
```

```python
us_weather=pd.DataFrame({
    'city':['New York','Los-Angeles','Chicago'],
    'temperature':[20,24,25],
    'humidity':[60,65,70]})
us_weather
```

```python
# Concat two data frames
c=pd.concat([india_weather,us_weather])
c
```

```python
# If we want continuous index
c=pd.concat([india_weather,us_weather],ignore_index=True)
c
```

```python
c=pd.concat([india_weather,us_weather], axis=0)
c
```

#### Merging of Data Frame

```python
temperature_df=pd.DataFrame({
    'city':['mumbai','bhubaneswar','delhi','bangalore'],
    'temperature':[29,34,30,25]})
temperature_df
```

```python
humidity_df=pd.DataFrame({
    'city':['mumbai','bhubaneswar','bangalore'],
    'humidity':[69,74,65]})
humidity_df
```

```python
# Merge two dataframes without explicitily mentioning index
df=pd.merge(temperature_df,humidity_df, on='city')
df
```

```python
# OUTER-JOIN
df=pd.merge(temperature_df,humidity_df, on='city', how='outer')
df
```

```python
# inner-join
df=pd.merge(temperature_df,humidity_df, on='city', how='inner')
df
```

### Logical operators

```python
# Creating a new subset using 'and' for multiple conditions.
filter1=df[(df['upvotes']<=10) & (df['tags']=='Chat')]
print('Shape of new dataset using and is:',filter1.shape)
```

```python
filter1
```

```python
# Creating a new subset using 'or' operator for multiple conditions.
filter2=df[(df['upvotes']<=10) | (df['tags']=='Chat') | (df['category']=='Productivity')]
print('Shape of new dataset using or is:',filter2.shape)
```

```python
filter2
```

```python
#Using both 'and' and 'or' together for multiple conditions.
filter3=df[(df['upvotes']<=10) | (df['tags']=='Chat') & (df['category']=='Productivity')]
print('Shape of new dataset using or is:',filter3.shape)
```
